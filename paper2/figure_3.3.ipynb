```python
from PIL import Image

# Load images
top = Image.open('/home/acker/WHIPS/ALA/paper2/fig3_3a.png')  # EPA minus Uncorrected
bottom = Image.open('/home/acker/WHIPS/ALA/paper2/fig3_3b.png')  # AQS minus CMAQ

# Resize to the same width (if needed)
width = max(top.width, bottom.width)
top = top.resize((width, int(top.height * width / top.width)))
bottom = bottom.resize((width, int(bottom.height * width / bottom.width)))

# Define overlap amount (in pixels)
overlap = 80  # move bottom image up by 100 pixels

# Total canvas height (subtract overlap)
total_height = top.height + bottom.height - overlap

# Create blank canvas
combined = Image.new("RGB", (width, total_height), (255, 255, 255))

# Paste top image first (background)
combined.paste(top, (0, 0))

# Paste bottom image next (foreground), slightly up
combined.paste(bottom, (0, top.height - overlap))

# Save the modified image

# Save the result
#combined.save('/home/acker/WHIPS/CMAQ/figure2.jpg', format='JPEG', quality=85, optimize=True)
```

```python
from PIL import Image

# Load images
left = Image.open('/home/acker/WHIPS/ALA/paper2/fig1a.png')   # EPA minus Uncorrected
right = Image.open('/home/acker/WHIPS/ALA/paper2/fig1b.png')  # AQS minus CMAQ

# Resize to same height (optional, ensures clean alignment)
height = max(left.height, right.height)
left = left.resize((int(left.width * height / left.height), height))
right = right.resize((int(right.width * height / right.height), height))

# Define overlap amount (horizontal)
overlap = 0  # in pixels

# Calculate total canvas width
total_width = left.width + right.width - overlap

# Create blank canvas
combined = Image.new("RGB", (total_width, height), (255, 255, 255))

# Paste left image
combined.paste(left, (0, 0))

# Paste right image, overlapping if needed
combined.paste(right, (left.width - overlap, 0))

# Save the result
# combined.save('/home/acker/WHIPS/ALA/paper2/figure1_combined.jpg', format='JPEG', quality=85, optimize=True)

combined
```
