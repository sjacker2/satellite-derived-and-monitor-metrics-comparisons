```python
import geopandas as gpd
df = gpd.read_file('/data/acker/ALA/paper2/all_variables.shp')
df
```

```python
df = df.rename(columns={'monitor__1':'monitor_coverage_pct', 'classifica':'classification', 'Design Val':'Design Value', 'monitor_co':'monitor_count',
                        'fire_regio':'fire_region', 'mountain_r':'mountain_region', 'desert_reg':'desert_region', 'urban_cate':'urban_category'})
```

```python
df
```

```python
# Replace values in the classification column to rename to alignment vs non-alignment
df['classification'] = df['classification'].replace({
    'TP': 'AP',
    'TN': 'AN',
    'FP': 'NP',
    'FN': 'NN'
})
```

```python
df.columns
```

```python
df['score'] = 0

# Add 1 point for each condition that is met
df['score'] += (df['monitor_count'] <= 2).astype(int)
df['score'] += (df['cdv_bin'].isin(['<7', '>10'])).astype(int)
df['score'] += ((df['monitor_coverage_pct'] < 0.05) | (df['monitor_coverage_pct'] > 0.1)).astype(int)
df['score'] += (df['fire_region'] == 'Fire').astype(int)
df['score'] += (df['mountain_region'] == 'Mountain').astype(int)
df['score'] += (df['desert_region'] == 'Desert').astype(int)
df['score'] += (df['urban_category'] == 'Non-Urban (<50%)').astype(int)
df['score'] += (df['size'] == 'large').astype(int)

# Sort by score descending
ranked = df.sort_values('score', ascending=False)
```

```python
ranked.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is called `df` and has columns 'abs_diff' and 'explanatory_variable_count'
sns.boxplot(x='score', y='abs_diff', data=ranked, color='lightgray')
sns.stripplot(x='score', y='abs_diff', data=ranked, color='black', jitter=True)

plt.xlabel('Number of Explanatory Variables')
plt.ylabel('Absolute Difference (µg/m³)')
plt.title('Relationship Between Explanatory Variables and Absolute Difference')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# First, create a new column for broader classification groups
ranked['agreement'] = ranked['classification'].apply(lambda x: 'Aligned (AP/AN)' if x in ['AP', 'AN'] else 'Non-Aligned (NP/NN)')

# Set the style
#sns.set(style="whitegrid")

# Boxplot showing distribution by number of explanatory variables, colored by agreement
sns.boxplot(
    x='score', y='abs_diff', hue='agreement', data=ranked,
    palette={'Aligned (AP/AN)': '#B4D4DF', 'Non-Aligned (NP/NN)': '#E28E8E'},
    dodge=True
)

# Overlay individual points
sns.stripplot(
    x='score', y='abs_diff', hue='agreement', data=ranked,
    dodge=True, jitter=True, marker='o', color='black', alpha=0.7
)
# Label axes and title
plt.xlabel('Number of Risk Factors Per County',fontsize=14)
plt.ylabel('Absolute Difference (µg/m³)', fontsize=14)
plt.rcParams["font.family"] = "serif"

# Fix the legend to avoid duplicates
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles[0:2], labels[0:2], title='Classification Group', fontsize=11)


plt.tight_layout()
plt.show()
```

```python
# Compute and print median absolute differences by score and agreement group
median_diffs = ranked.groupby(['score', 'agreement'])['abs_diff'].median().reset_index()
print(median_diffs)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# --- Add separate scores for each group ---
ranked['score_all'] = 0
ranked['score_monitor'] = 0
ranked['score_geo'] = 0

# All risk factors
ranked['score_all'] += (ranked['monitor_count'] <= 2).astype(int)
ranked['score_all'] += (ranked['cdv_bin'].isin(['<7', '>10'])).astype(int)
ranked['score_all'] += ((ranked['monitor_coverage_pct'] < 0.05) | (ranked['monitor_coverage_pct'] > 0.1)).astype(int)
ranked['score_all'] += (ranked['fire_region'] == 'Fire').astype(int)
ranked['score_all'] += (ranked['mountain_region'] == 'Mountain').astype(int)
ranked['score_all'] += (ranked['desert_region'] == 'Desert').astype(int)
ranked['score_all'] += (ranked['urban_category'] == 'Non-Urban (<50%)').astype(int)
ranked['score_all'] += (ranked['size'] == 'large').astype(int)

# Monitor-related only
ranked['score_monitor'] += (ranked['monitor_count'] <= 2).astype(int)
ranked['score_monitor'] += (ranked['cdv_bin'].isin(['<7', '>10'])).astype(int)
ranked['score_monitor'] += ((ranked['monitor_coverage_pct'] < 0.05) | (ranked['monitor_coverage_pct'] > 0.1)).astype(int)

# Geography-related only
ranked['score_geo'] += (ranked['fire_region'] == 'Fire').astype(int)
ranked['score_geo'] += (ranked['mountain_region'] == 'Mountain').astype(int)
ranked['score_geo'] += (ranked['desert_region'] == 'Desert').astype(int)
ranked['score_geo'] += (ranked['urban_category'] == 'Non-Urban (<50%)').astype(int)
ranked['score_geo'] += (ranked['size'] == 'large').astype(int)

# Broader classification groups
ranked['agreement'] = ranked['classification'].apply(
    lambda x: 'Aligned (AP/AN)' if x in ['AP', 'AN'] else 'Non-Aligned (NP/NN)'
)

# --- Plot: 3 Panels ---
fig, axes = plt.subplots(3, 1, figsize=(10, 18), sharey=True)
sns.set(font='serif', style="whitegrid", font_scale=1.2)

# Color palette
palette = {'Aligned (AP/AN)': '#B4D4DF', 'Non-Aligned (NP/NN)': '#E28E8E'}

# Panel a: All risk factors
sns.boxplot(
    x='score_all', y='abs_diff', hue='agreement', data=ranked,
    palette=palette, dodge=True, ax=axes[0]
)
'''sns.stripplot(
    x='score_all', y='abs_diff', hue='agreement', data=ranked,
    dodge=True, jitter=True, marker='o', color='black', alpha=0.6, ax=axes[0]
)'''
axes[0].set_title("(a) All Risk Factors", fontsize=17)
axes[0].set_xlabel('', fontsize=14)
axes[0].set_ylabel('Absolute Difference (µg/m³)', fontsize=15)
axes[0].legend_.remove()

# Panel b: Monitor-related risk factors
sns.boxplot(
    x='score_monitor', y='abs_diff', hue='agreement', data=ranked,
    palette=palette, dodge=True, ax=axes[1]
)
'''sns.stripplot(
    x='score_monitor', y='abs_diff', hue='agreement', data=ranked,
    dodge=True, jitter=True, marker='o', color='black', alpha=0.6, ax=axes[1]
)'''
axes[1].set_title("(b) Monitor-Related Risk Factors", fontsize=17)
axes[1].set_xlabel('', fontsize=15)
axes[1].set_ylabel('Absolute Difference (µg/m³)', fontsize=15)
axes[1].legend_.remove()

# Panel c: Geography-related risk factors
sns.boxplot(
    x='score_geo', y='abs_diff', hue='agreement', data=ranked,
    palette=palette, dodge=True, ax=axes[2]
)
'''sns.stripplot(
    x='score_geo', y='abs_diff', hue='agreement', data=ranked,
    dodge=True, jitter=True, marker='o', color='black', alpha=0.6, ax=axes[2]
)'''
axes[2].set_title("(c) Geography-Related Risk Factors", fontsize=17)
axes[2].set_xlabel('Number of Risk Factors', fontsize=15)
axes[2].set_ylabel('Absolute Difference (µg/m³)', fontsize=15)
axes[2].legend_.remove()

# Fix legend (single legend below the panels)
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(
    handles[:2], labels[:2],
    title= None, fontsize=14,
    loc='lower center', bbox_to_anchor=(0.5, -0.02), ncol=2, frameon=False
)

plt.tight_layout()
plt.show()
```
