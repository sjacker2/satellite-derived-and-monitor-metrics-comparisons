```python
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.patches as mpatches
import matplotlib.lines as mlines


# Load the images
img1 = Image.open("/home/acker/WHIPS/ALA/paper2/diff_TP.png")
img2 = Image.open("/home/acker/WHIPS/ALA/paper2/diff_TN.png")
img3 = Image.open("/home/acker/WHIPS/ALA/paper2/diff_FP.png")
img4 = Image.open("/home/acker/WHIPS/ALA/paper2/diff_FN.png")

# Use serif font for all text
plt.rcParams["font.family"] = "serif"

# Titles for each panel
'''titles = [
    "a) EPA CDVs",
    "b) Monitored CDVEs",
    "c) Unmonitored CDVEs",
    "d) All U.S. CDVEs"
]'''

# Images for each panel
images = [img1, img2, img3, img4]

# Define the color range for the horizontal legend
range_colors = {
    'AP': "#00A86B",
    'AN': "#8E0EA5",
    'NP': "#ACDF87",
    'NN': "#fe9bfe"
}

# Create the figure with controlled spacing
fig, axes = plt.subplots(2, 2, figsize=(16, 13), gridspec_kw={'wspace': -0.005, 'hspace': -0.35})

# Plot each image
for ax, img in zip(axes.flatten(), images):
    ax.imshow(img)
    #ax.set_title(title, fontsize=20, fontfamily='serif', pad=0)
    ax.axis("off")

# Create legend patches in correct order for horizontal layout
legend_labels = ["AP", "AN", "NP", "NN"]

# Create circular legend handles
legend_patches = [mpatches.Patch(facecolor=range_colors[label], edgecolor="black" if label == "N/A" else "none", label=label)
                  for label in legend_labels]


# Add the horizontal legend below all plots
fig.legend(
    handles=legend_patches,
    loc='lower center',
    ncol=4,
    fontsize=11,
    frameon=False,
    bbox_to_anchor=(0.52, 0.17)
)

plt.rcParams["font.family"] = "serif"

#plt.tight_layout()
plt.subplots_adjust(bottom=0.12)  # Make space for legend
plt.show()
```
