```python
df = gpd.read_file('/data/acker/ALA/paper2/washu_2021-2023_merged.shp')
df
```

```python
merged_fires = gpd.sjoin(fires, df, how='inner', predicate='intersects')
merged_fires
```

```python
# Create a column in mountain counties to indicate they are in the mountain region
merged_fires_regions =merged_fires[['GEOID']].copy()
merged_fires_regions['fire_region'] = 'Fire'

# Merge with all counties, marking non-mountain counties
all_counties_gdf = df.merge(merged_fires_regions, on='GEOID', how='left')

# Fill in missing values as 'Non-Mountain'
all_counties_gdf['fire_region'] = all_counties_gdf['fire_region'].fillna('Non-Fire')
```

```python
all_counties_gdf = all_counties_gdf.drop_duplicates(subset='GEOID')
```

```python
all_counties_gdf.to_file('/data/acker/ALA/paper2/fire_counties.shp')
```

```python
all_counties_gdf['diff'] = all_counties_gdf['PM25_90th'] - all_counties_gdf['Design Val']
all_counties_gdf['abs_diff'] = abs(all_counties_gdf['PM25_90th'] - all_counties_gdf['Design Val'])
```

```python
f = all_counties_gdf[all_counties_gdf['fire_region'] == 'Fire']
f['abs_diff'].describe()
```

```python
f = all_counties_gdf[all_counties_gdf['fire_region'] == 'Non-Fire']
f['abs_diff'].describe()
```

```python
# Define threshold
threshold = 9.0

# Function to classify each row
def classify(row):
    cdv = row['Design Val']
    cdve = row['PM25_90th']
    if cdv > threshold and cdve > threshold:
        return 'AP'
    elif cdv > threshold and cdve <= threshold:
        return 'NP'
    elif cdv <= threshold and cdve > threshold:
        return 'NN'
    else:
        return 'AN'

# Apply the classification
all_counties_gdf['classification'] =all_counties_gdf.apply(classify, axis=1)

# Print counts
print(all_counties_gdf['classification'].value_counts())

# Optional: filter into separate DataFrames
AP =all_counties_gdf[all_counties_gdf['classification'] == 'AP']
NP =all_counties_gdf[all_counties_gdf['classification'] == 'NP']
NN =all_counties_gdf[all_counties_gdf['classification'] == 'NN']
AN =all_counties_gdf[all_counties_gdf['classification'] == 'AN']

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 6))
plt.rcParams["font.family"] = "serif"


# Create the boxplot
sns.boxplot(
    data=all_counties_gdf,
    x='fire_region',
    y='diff',
    hue='classification',
    palette={'AP': '#83b5fa', 'AN': '#074cad', 'NP': '#fe9bfe', 'NN': '#8E0EA5'}
)


# Style and formatting
plt.axhline(0, color='black', linestyle='--')
plt.xlabel('', fontsize=13)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel('Difference (CDVE – CDV) (µg/m³)', fontsize=16)
plt.title('(a)', fontsize=18)
plt.legend().remove()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
# Group by classification and mountain region
stats = (
    all_counties_gdf
    .groupby(['classification', 'fire_region'])['diff']
    .agg(['count', 'mean', 'std', 'min', 'max', 'median'])
    .reset_index()
    .sort_values(by=['classification', 'fire_region'])
)

# Add a new column for range
stats['range'] = stats['max'] - stats['min']

# Display the summary
pd.set_option('display.float_format', '{:.2f}'.format)
print(stats)
```

```python
# Assuming you've already done the spatial join and have a GeoDataFrame like:
# counties_with_fires = gpd.sjoin(county_gdf, fire_gdf, how="inner", predicate="intersects")

# Now group by county GEOID and sum the burned area
burned_area_by_county = (
    merged_fires
    .groupby('GEOID')['BurnBndAc']
    .sum()
    .reset_index()
    .rename(columns={'BurnBndAc': 'total_burned_acres'})
)

# Preview the result
print(burned_area_by_county.head())
```

```python
# Join back to full county GeoDataFrame (to add the total burned acres column)
merged_df = df.merge(burned_area_by_county, on='GEOID', how='left')

# Fill NaNs for counties with no fires
merged_df['total_burned_acres'] = merged_df['total_burned_acres'].fillna(0)
```

```python
merged_df = merged_df.rename(columns={'Design Val':'Design Value'})
```

```python
merged_df['diff'] =  merged_df['PM25_90th'] - merged_df['Design Value']
```

```python
# Define threshold
threshold = 9.0

# Function to classify each row
def classify(row):
    cdv = row['Design Value']
    cdve = row['PM25_90th']
    if cdv > threshold and cdve > threshold:
        return 'AP'
    elif cdv > threshold and cdve <= threshold:
        return 'NP'
    elif cdv <= threshold and cdve > threshold:
        return 'NN'
    else:
        return 'AN'

# Apply the classification
merged_df['classification'] = merged_df.apply(classify, axis=1)

# Print counts
print(merged_df['classification'].value_counts())

# Optional: filter into separate DataFrames
AP = merged_df[merged_df['classification'] == 'AP']
NP = merged_df[merged_df['classification'] == 'NP']
NN = merged_df[merged_df['classification'] == 'NN']
AN = merged_df[merged_df['classification'] == 'AN']
```

```python
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

# Define color mapping for classification
colors = {
    'AP': "#00A86B",
    'AN': "#8E0EA5",
    'NP': "#ACDF87",
    'NN': "#fe9bfe"
}

plt.figure(figsize=(8, 6))

# Scatter plot grouped by classification
for label, group in merged_df.groupby('classification'):
    plt.scatter(
        group['total_burned_acres'],  # x-axis
        group['diff'],                # y-axis
        alpha=0.7,
        edgecolors=None,
        label=label,
        color=colors.get(label, 'gray')
    )

# Add horizontal line at y = 0
plt.axhline(0, linestyle='--', color='black')

# Apply log scale to the x-axis
plt.xscale('log')

# Set custom ticks and labels for burned acres
xticks = [500, 1000, 10000, 100000, 1000000]
xtick_labels = ['500', '1k', '10k', '100k', '1M']
plt.xticks(xticks, xtick_labels)

# Labels and formatting
plt.rcParams["font.family"] = "serif"
plt.xlabel('Area Burned Per County 2021–2023 (Acres)', fontsize=13)
plt.ylabel('Difference (CDVE – CDV) (µg/m³)', fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(False)

# Horizontal legend below plot
legend_order = ['AP', 'AN', 'NP', 'NN']
legend_handles = [
    mlines.Line2D([], [], color=colors[label], marker='o', linestyle='None',
                  markersize=7, markeredgecolor=None, label=label)
    for label in legend_order
]

plt.legend(
    handles=legend_handles,
    loc='lower center',
    bbox_to_anchor=(0.5, -0.22),
    ncol=4,
    frameon=False,
    fontsize=12
)

plt.tight_layout()
plt.show()
```
