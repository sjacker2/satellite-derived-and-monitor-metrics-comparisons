```python
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.patches as mpatches
import matplotlib.lines as mlines


# Load the images
img1 = Image.open("/home/acker/WHIPS/ALA/paper2/wildfires.png")
img2 = Image.open("/home/acker/WHIPS/ALA/paper2/mountains.png")
img3 = Image.open("/home/acker/WHIPS/ALA/paper2/desert.png")
img4 = Image.open("/home/acker/WHIPS/ALA/paper2/urban.png")

# Use serif font for all text
plt.rcParams["font.family"] = "serif"

# Titles for each panel
'''titles = [
    "a) EPA CDVs",
    "b) Monitored CDVEs",
    "c) Unmonitored CDVEs",
    "d) All U.S. CDVEs"
]'''

# Images for each panel
images = [img1, img2, img3, img4]

# Define the color range for the horizontal legend
range_colors = {
    'TP': "#83b5fa",
    'TN': '#074cad',
    'FP': "#fe9bfe",
    'FN': "#8E0EA5"
}

# Create the figure with controlled spacing
fig, axes = plt.subplots(2, 2, figsize=(16, 13), gridspec_kw={'wspace': -0.005, 'hspace': -0.3})

# Plot each image
for ax, img in zip(axes.flatten(), images):
    ax.imshow(img)
    #ax.set_title(title, fontsize=20, fontfamily='serif', pad=0)
    ax.axis("off")

# Create legend patches in correct order for horizontal layout
legend_labels = ["FP", "FN", "TP", "TN"]

# Create circular legend handles
legend_patches = [mpatches.Patch(facecolor=range_colors[label], edgecolor="black" if label == "N/A" else "none", label=label)
                  for label in legend_labels]


# Add the horizontal legend below all plots
fig.legend(
    handles=legend_patches,
    loc='lower center',
    ncol=4,
    fontsize=11,
    frameon=False,
    bbox_to_anchor=(0.52, 0.15)
)

plt.rcParams["font.family"] = "serif"

#plt.tight_layout()
plt.subplots_adjust(bottom=0.12)  # Make space for legend
plt.show()
```
