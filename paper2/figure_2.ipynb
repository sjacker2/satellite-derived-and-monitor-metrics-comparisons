```python
from PIL import Image

# Load images
img1 = Image.open('/home/acker/WHIPS/ALA/paper2/fig2a.png')
img2 = Image.open('/home/acker/WHIPS/ALA/paper2/fig2b.png')
img3 = Image.open('/home/acker/WHIPS/ALA/paper2/fig2c.png')

# Resize all images to the same width (if needed)
width = max(img1.width, img2.width, img3.width)

img1 = img1.resize((width, int(img1.height * width / img1.width)))
img2 = img2.resize((width, int(img2.height * width / img2.width)))
img3 = img3.resize((width, int(img3.height * width / img3.width)))

# Optional overlap between images (set to 0 for no overlap)
overlap = 0  

# Calculate total height for canvas
total_height = img1.height + img2.height + img3.height - 2 * overlap

# Create a blank canvas
combined = Image.new("RGB", (width, total_height), (255, 255, 255))

# Paste images on canvas
combined.paste(img1, (0, 0))
combined.paste(img2, (0, img1.height - overlap))
combined.paste(img3, (0, img1.height + img2.height - 2 * overlap))

# Save the combined image
combined.save('/home/acker/WHIPS/ALA/paper2/fig2_combined.png', format='PNG', optimize=True)

# Optionally, show the result
combined.show()
```
