```python
import geopandas as gpd
df = gpd.read_file('/data/acker/ALA/paper2/all_variables.shp')
df
```

```python
df = df.rename(columns={'monitor__1':'monitor_coverage_pct', 'classifica':'classification', 'Design Val':'Design Value', 'monitor_co':'monitor_count',
                        'fire_regio':'fire_region', 'mountain_r':'mountain_region', 'desert_reg':'desert_region', 'urban_cate':'urban_category'})
```

```python
df
```

```python
df.columns
```

```python
df = df[(df['classification'] == 'FP') | (df['classification'] == 'FN')]
```

```python
df['score'] = 0

# Add 1 point for each condition that is met
df['score'] += (df['monitor_count'] <= 2).astype(int)
df['score'] += (df['cdv_bin'].isin(['<7', '>10'])).astype(int)
df['score'] += ((df['monitor_coverage_pct'] >= 0.05) & (df['monitor_coverage_pct'] < 0.1)).astype(int)
df['score'] += (df['fire_region'] == 'Fire').astype(int)
df['score'] += (df['mountain_region'] == 'Mountain').astype(int)
df['score'] += (df['desert_region'] == 'Desert').astype(int)
df['score'] += (df['urban_category'] == 'Non-Urban (<50%)').astype(int)

# Sort by score descending
ranked = df.sort_values('score', ascending=False)
```
