```python
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.patches as mpatches

# Load the images
img1 = Image.open("/home/acker/WHIPS/ALA/GL/epa_dv.jpg")
img2 = Image.open("/home/acker/WHIPS/ALA/GL/monitor_CDVE.jpg")
img3 = Image.open("/home/acker/WHIPS/ALA/GL/unmon_CDVE.jpg")
img4 = Image.open("/home/acker/WHIPS/ALA/GL/all_cdves.jpg")

# Use serif font for all text
plt.rcParams["font.family"] = "serif"

# Titles for each panel
titles = [
    "a) EPA CDVs",
    "b) Monitored CDVEs",
    "c) Unmonitored CDVEs",
    "d) All U.S. CDVEs"
]

# Images for each panel
images = [img1, img2, img3, img4]

# Define the color range for the horizontal legend
range_colors = {
    "< 4 g/m": "#ffa600",
    "4 - 6 g/m": "#ff764a",
    "6 - 7 g/m": "#ef5675",
    "7 - 8 g/m": "#bc5090",
    "8 - 9 g/m": "#7a5195",
    "9 - 12 g/m": "#374c80",
    "> 12 g/m": "#003f4c",
    "N/A": "white"
}

# Create the figure with controlled spacing
fig, axes = plt.subplots(2, 2, figsize=(16, 13), gridspec_kw={'wspace': -0.005, 'hspace': -0.2})

# Plot each image
for ax, img, title in zip(axes.flatten(), images, titles):
    ax.imshow(img)
    ax.set_title(title, fontsize=20, fontfamily='serif', pad=0)
    ax.axis("off")

# Create legend patches in correct order for horizontal layout
legend_labels = ["< 4 g/m", "4 - 6 g/m", "6 - 7 g/m", "7 - 8 g/m",
                 "8 - 9 g/m", "9 - 12 g/m", "> 12 g/m", "N/A"]
legend_patches = [mpatches.Patch(facecolor=range_colors[label], edgecolor="black" if label == "N/A" else "none", label=label)
                  for label in legend_labels]

# Add the horizontal legend below all plots
fig.legend(
    handles=legend_patches,
    loc='lower center',
    ncol=4,
    fontsize=16,
    frameon=False,
    bbox_to_anchor=(0.5, 0.13)
)

#plt.tight_layout()
plt.subplots_adjust(bottom=0.12)  # Make space for legend
plt.show()
```
