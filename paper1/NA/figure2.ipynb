```python
import matplotlib.pyplot as plt
from PIL import Image

# Load the images
img1 = Image.open("/home/acker/WHIPS/ALA/NA/epa_dv.jpg")  # Replace with correct paths if necessary
img2 = Image.open("/home/acker/WHIPS/ALA/NA/monitor_CDVE.jpg")
img3 = Image.open("/home/acker/WHIPS/ALA/NA/unmon_CDVE.jpg")
img4 = Image.open("/home/acker/WHIPS/ALA/NA/all_cdves.jpg")

# Create the figure with controlled spacing
fig, axes = plt.subplots(
    2, 2, figsize=(16, 12),
    gridspec_kw={'wspace': -0.005, 'hspace': -0.2}  # Adjust spacing between subplots
)

# Load custom font
plt.rcParams["font.family"] = "serif"

# Titles for each panel
titles = [
    "a) EPA CDVs",
    "b) Monitored CDVEs",
    "c) Unmonitored CDVEs",
    "d) All U.S. CDVEs"
]

# Images for each panel
images = [img1, img2, img3, img4]

# Plot each image
for ax, img, title in zip(axes.flatten(), images, titles):
    ax.imshow(img)
    ax.set_title(title, fontsize=20, pad=0)  # Reduce padding to bring title closer to the image
    ax.axis("off")  # Turn off axis labels


plt.show()
```
